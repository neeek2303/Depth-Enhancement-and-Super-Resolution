# 2. Render loop

```python
import glob
import tqdm 
folders = sorted(glob.glob(f'{output_path}/*'))
```

```python
import trimesh
import pyrender


def render_depth(mesh, pose_path, intrinsic, scale = 2, width = 640, height = 480):

    r = pyrender.OffscreenRenderer(scale*width, scale*height)
    scene = pyrender.Scene()

    # https://answers.opencv.org/question/23089/opencv-opengl-proper-camera-pose-using-solvepnp/
    # https://amytabb.com/ts/2019_06_28/
    # Y axis and Z axis are inverted.
    # z-axis away from the scene, x-axis right, y-axis up
    refl = np.eye(4,4)
    refl[2,2] = -1
    refl[1,1] = -1

    

    camera_pose = np.loadtxt(pose_path)
    camera_pose = camera_pose@refl

    # call this only per scene, it should not be needed more than once per scene
    fx,fy,cx,cy = [i*scale for i in intrinsic]
    camera = pyrender.IntrinsicsCamera(fx,fy,cx,cy)
    scene.clear() # clear all meshes, cameras and poses

    scene.add(mesh)
    camera_pose_node = scene.add(camera, pose=camera_pose)
    color, depth = r.render(scene)
    return depth
```

```python
np.random.seed(23)
val_folders = np.random.choice(len(folders), int(len(folders)*0.05),replace=False)
```

```python
len(val_folders), len(folders)
```

```python
def find_numb(pose_path):
    return int(pose_path.split('/')[-1].split('.')[0])
```

```python
save_folder = '/home/Scannet_data/'
```

```python
v = 0
t = 0
test = 0 
t_te=0
for indx, fold in tqdm(enumerate(folders)):
#     print(fold)
    images = sorted(glob.glob(fold+'/color/*'))
    depthes = sorted(glob.glob(fold+'/depth/*'))
    poses = sorted(glob.glob(fold+'/pose/*'))
    intr = np.loadtxt(fold +"/intrinsic/intrinsic_depth.txt")
    intrinsic = (intr[0,0], intr[1,1], intr[0,2], intr[1,2])
    pose_list = list(map(find_numb, poses))
    f = fold.split('/')[-1]  

    ply_path  = path_to_scans +f +'/'+ f+ '_vh_clean.ply'
    mesh = pyrender.Mesh.from_trimesh(trimesh.load_mesh(ply_path))
    
    if indx in val_folders:
        for i in range(len(images)):
            v+=1
            im = imageio.imread(images[i])
            d = imageio.imread(depthes[i])
            d_p = int(depthes[i].split('_')[-1].split('.')[0])
            p = poses[pose_list.index(d_p)]
            try:
                d_gt = render_depth(mesh, p, intrinsic)
                imageio.imwrite(f'{save_folder}images_val/'+images[i].split('/')[-1], im)
                imageio.imwrite(f'{save_folder}depths_val/'+depthes[i].split('/')[-1], d)
                imageio.imwrite(f'{save_folder}depths_val_gt/'+depthes[i].split('/')[-1], (d_gt*1000).astype(np.uint16))
            except Exception:
                with open(f'{save_folder}exeptions.txt', 'a', encoding='utf-8') as g:
                    print('index: {}, fold: {},    \n'.format(indx,fold), file=g)
                print(Exception)

    else:
        np.random.seed(23)
        indexes_train = np.random.choice(len(images), len(images)*4//5, replace=False)
#         print(len(indexes_train), len(images),len(indexes_train)/len(images))
#         print(sorted(indexes_train))
        for i in range(len(images)):
            im = imageio.imread(images[i])
            d = imageio.imread(depthes[i])
            d_p = int(depthes[i].split('_')[-1].split('.')[0])
            p = poses[pose_list.index(d_p)]
            try:
                d_gt = render_depth(mesh, p, intrinsic)
#                 t_te+=1
#                 print('index: ', i)
                if i in indexes_train:
#                     t+=1
#                     print('train: ', i)
                    imageio.imwrite(f'{save_folder}/images_train/'+images[i].split('/')[-1], im)
                    imageio.imwrite(f'{save_folder}/depths_train/'+depthes[i].split('/')[-1], d)
                    imageio.imwrite(f'{save_folder}/depths_train_gt/'+depthes[i].split('/')[-1], (d_gt*1000).astype(np.uint16))                    
                else:
#                     test+=1
#                     print('test: ', i)
                    imageio.imwrite(f'{save_folder}/images_test/'+images[i].split('/')[-1], im)
                    imageio.imwrite(f'{save_folder}/depths_test/'+depthes[i].split('/')[-1], d)
                    imageio.imwrite(f'{save_folder}/depths_test_gt/'+depthes[i].split('/')[-1], (d_gt*1000).astype(np.uint16))
            except Exception:
                with open(f'{save_folder}/exeptions.txt', 'a', encoding='utf-8') as g:
                    print('index: {}, fold: {},    \n'.format(indx,fold), file=g)
                print(Exception)
```

```python
def get_normal(depth):
    norm = np.zeros((3, depth.shape[0], depth.shape[1]))
    dzdx = np.gradient(depth, 1, axis=0)
    dzdy = np.gradient(depth, 1, axis=1)
    norm[0, :, :] = -dzdx
    norm[1, :, :] = -dzdy
    norm[2, :, :] = np.ones_like(depth)
    n = np.linalg.norm(norm, axis = 0, keepdims=True)
    norm = norm/n
    return norm
```
