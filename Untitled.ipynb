```python
import matplotlib.pyplot as plt
import os
import imageio
```

```python
import numpy as np
```

```python
import albumentations as A
def apply_transformer(transformations,  depth):
    res = A.Compose(transformations, p=1)(image=depth)
    return res
```

```python
def get_normal(depth):
    norm = np.zeros(( depth.shape[0], depth.shape[1], 3))
    dzdx = np.gradient(depth, 1, axis=0)
    dzdy = np.gradient(depth, 1, axis=1)
    norm[ :, :, 0] = -dzdx
    norm[ :, :, 1] = -dzdy
    norm[ :, :, 2] = np.ones_like(depth)
    n = np.linalg.norm(norm, axis = 2, ord=2, keepdims=True)
    norm = norm/(n + 1e-15)
    return norm
```
